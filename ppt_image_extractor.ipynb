```python
# Import Required Libraries
from pathlib import Path
from PIL import Image
from pptx import Presentation
from pptx.enum.shapes import PP_PLACEHOLDER
import os
import io
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
# Load the PowerPoint File
ppt_path = Path('Mentor_Dashboard_Report.pptx')
if not ppt_path.exists():
    raise FileNotFoundError(f'PowerPoint file not found: {ppt_path}')

prs = Presentation(str(ppt_path))
print(f'Loaded presentation: {ppt_path}')
print(f'Total slides: {len(prs.slides)}')
```

```python
# Iterate Through Slides
for i, slide in enumerate(prs.slides):
    print(f'Slide {i+1}: shapes={len(slide.shapes)}')
```

```python
# Extract Images from Slides
output_dir = Path('ppt_images')
output_dir.mkdir(exist_ok=True)

image_count = 0
for slide_idx, slide in enumerate(prs.slides):
    for shape_idx, shape in enumerate(slide.shapes):
        if hasattr(shape, 'image'):
            image_count += 1
            image_bytes = shape.image.blob
            image = Image.open(io.BytesIO(image_bytes))
            img_path = output_dir / f'slide_{slide_idx + 1}_image_{shape_idx + 1}.png'
            image.save(img_path)
            print(f'Saved: {img_path}')

print(f'Total extracted images: {image_count}')
```

```python
# Save Images to the Output Folder
for file in sorted(output_dir.glob('*')):
    print(file)
```

```python
# Display Extracted Images
from IPython.display import Image as IPImage

for image_file in sorted(output_dir.glob('*.png')):
    print(f'Previewing: {image_file}')
    display(IPImage(filename=str(image_file), width=600))
```
